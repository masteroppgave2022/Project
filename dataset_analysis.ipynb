```python
from data.preprocess_functions import Preprocess
import matplotlib.pyplot as plt 
import numpy as np
import rasterio
import pandas as pd
import os
```

```python
folder = '/localhome/studenter/renatask/Project/data/unprocessed_downloads'
pp = Preprocess()

widths = []
heights = []
names = []
band_namess = []
orbits = []
looks = []
product_types =[]
descriptors = []
acquisition_modes = []
first_line_times = []
last_line_times = []
pulse_repetition_frequencys = []
radar_frequencys =  []
total_sizes =  []

len = len(os.listdir(folder))
count = 0

for file in os.listdir(folder):
    if file.startswith('.'): continue
    if count == 26: 
        count +=1 
        continue
    count += 1
    product = pp.read_product(folder+'/'+file)

    width = product.getSceneRasterWidth()
    height = product.getSceneRasterHeight()
    name = product.getName()
    band_names = product.getBandNames()
    orbit = product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('PASS').getData()
    look = product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('antenna_pointing').getData()
    product_type = product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('PRODUCT_TYPE').getData()
    descriptor = product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('SPH_DESCRIPTOR').getData()
    acquisition_mode = product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('ACQUISITION_MODE').getData()
    first_line_time = product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('first_line_time').getData()
    last_line_time = product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('last_line_time').getData()
    pulse_repetition_frequency =  product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('pulse_repetition_frequency').getData()
    radar_frequency =  product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('radar_frequency').getData()
    total_size =  product.getMetadataRoot().getElement('Abstracted_Metadata').getAttribute('total_size').getData()

    product = pp.read_product(folder +'/'+file)
    widths.append(width)
    heights.append(height)
    names.append(str(name)) #[0:32]
    band_namess.append("{}".format(", ".join(band_names)))
    orbits.append(str(orbit))
    looks.append(str(look))
    product_types.append(product_type)
    descriptors.append(descriptor)
    acquisition_modes.append(acquisition_mode)
    first_line_times.append(str(first_line_time)[0:20])
    last_line_times.append(last_line_time)
    pulse_repetition_frequencys.append(str(pulse_repetition_frequency)[0:9])
    radar_frequencys.append(str(radar_frequency)[0:6])
    total_sizes.append(total_size)

    


```

```python
data = {
    'Name': names,
    'Width': widths,
    'Height': heights,
    'Orbit': orbits,
    'Look': looks,
    'Bands': band_namess,
    'Product_Type': product_types,
    'Descriptor': descriptors,
    'Acqusition_Mode': acquisition_modes,
    'First_Line_Time': first_line_times,
    'Last_Line_Time': last_line_times,
    'Pulse_Repetition_Frequency': pulse_repetition_frequencys,
    'Radar_Frequency': radar_frequencys,
    'Total_Size': total_sizes
}

collection = pd.DataFrame(data)

collection.to_csv('dataset_description.csv')
collection.to_excel('dataset_description.xlsx')
collection.to_latex('dataset_description.tex')
```
